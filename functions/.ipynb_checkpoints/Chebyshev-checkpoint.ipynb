```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sp
from numpy.polynomial import chebyshev
import scipy.signal as sig
theta=np.linspace(-np.pi, np.pi, 100)
import pickle
```

Begin with the functions used for Hamiltonian Simulation in Gily√©n, Martyn: Jacobi-Anger expansions of cosine and sine

```python
def JA_COEF(n, t):
    'checked some manual values: signs and cos/sin assignments are ???'
    coscoeff, sincoeff, signs = np.zeros(n+1), np.zeros(n+1), np.ones(n+1)
    #signs[::4] = 1
    #signs[1::4]=1
    signs[2::4]=-1
    signs[3::4]=-1
    bessel=signs*2*sp.jv(np.arange(0, n+1, 1), t)
    bessel[0]=sp.jv(0, t)
    coscoeff[::2]=bessel[::2]
    sincoeff[1::2]=bessel[1::2]
    return coscoeff, sincoeff, bessel

def n_FIND(t, epsi, alpha=1, e=0.5):
    r=2*alpha*abs(t)+3*np.log(12/epsi)
    return int(np.ceil(2*r))
```

```python
def LAUR_POLY_MULT(coeffa, coeffb):
    """produces the coefficient list of the prodcut of polynomials a, b from their coefficient lists.
    The for loops are pretty sloppy but it doesn't matter. 
    Assumes a, b are Laurent polynomials of the same degree n. Would be easy to change, but doesn't matter.
    checked length or arrays, coeffc indexing, plus an n=1 poly. Others look convincing"""
#     coeffc=np.zeros(4*n+1,dtype=complex)
#     for k in range(-2*n,2*n+1):
#         ck=0
#         for i in range(-n, n+1):
#             if (k-i)>=-n and (k-i)<=n:
#                 ck=ck+coeffa[i+n]*coeffb[k-i+n]
#         coeffc[k+2*n]=ck
    return sig.convolve(coeffa, coeffb, mode="full", method='fft')

def LAUR_POLY_BUILD(coeff, n, z):
    """computes the float-point value of a Laurent polynomial from its coefficient list (does not assume symmetric coefficents).
    x should be scalar or array but it must NOT be integer-valued"""
    """gives the correct expression for some simple, nonsymmetric n=2 Laurents"""
    polyval=0
    for l in range(-n,n+1):
        polyval=polyval+coeff[l+n]*z**l
    return polyval

def FOUR_TO_LAUR(coscoeff, sincoeff):
    a=np.append(np.flip(coscoeff[1:]),coscoeff)/2
    b=np.append(-np.flip(sincoeff[1:]),sincoeff)/2j
    return a, b


def SENSIBLE_MATRIX(A, tol=10**(-15)):
    """function to make reading matrix results easier. checked against a simple example"""
    Ar=np.where(abs(np.real(A))>tol, np.real(A), 0)
    Ai=np.where(abs(np.imag(A))>tol, np.imag(A), 0)
    return Ar+Ai*1j
```
